```python
# Dependencies and Setup
import pandas as pd
import pathlib
import csv
import scipy.stats as st
import numpy as np
import os
```

```python
# Create paths for csv
edd_path = "../cleaned_data/by_county_edd_df.csv"
hospital_rating_path = "../cleaned_data/by_county_hospitalrating_ver3_df.csv"
zillow_path = "../cleaned_data/by_county_zillow_pricing_df.csv"
nu_hospitals_path = "../cleaned_data/by_county_count_hospital.csv"
zip_code_path = "../cleaned_data/zip_county.csv"
lnglat_CA_path = "../cleaned_data/lnglat_CA.csv"
```

```python
# Read csv
edd_df = pd.read_csv(edd_path)
# pd.DataFrame(edd_df)
```

```python
# Delete extraneous columns
new_edd_df = edd_df.drop(columns= [
    "laborForce"
    , "employment"
    , "unemployment"
])    
# new_edd_df
```

```python
# Read csv
hospital_rating_df = pd.read_csv(hospital_rating_path)
# pd.DataFrame(hospital_rating_df)
```

```python
# Read csv
zillow_df = pd.read_csv(zillow_path)
# pd.DataFrame(zillow_df)
```

```python
# Read csv
nu_hospitals_df = pd.read_csv(nu_hospitals_path)
# pd.DataFrame(nu_hospitals_df)
```

```python
# Rename columns
final_nu_hospitals_df = nu_hospitals_df.rename(columns={"Unnamed: 0":"countyName"})
# final_nu_hospitals_df
```

```python
# Read csv
zip_code_df = pd.read_csv(zip_code_path)
# zip_code_df
```

```python
# Zip code 
z = zip_code_df.set_index('countyName') 
z
```

```python
# Read csv
lnglat_CA_df = pd.read_csv(lnglat_CA_path)
pd.DataFrame(lnglat_CA_df)
```

```python
# edd + hospital ratings combination = a
a = new_edd_df.set_index('countyName').join(hospital_rating_df.set_index('countyName'))
a
```

```python
# count of hospitals + a = b
b = a.join(final_nu_hospitals_df.set_index('countyName'))
b
```

```python
# zillow + b = c
c = b.join(zillow_df.set_index('countyName'))
c
```

```python
# zip + c = d 
d = c.merge(z, how='inner', on='countyName')
d
```

```python
# d + c = d 
e = d.merge(lnglat_CA_df, how='inner', on='zip')
e
```

```python
# Fill 0 for all NaN for countOfHospital and hospitalOverallRating
f = e.fillna(0)
f
```

```python
# Final cleaned csv to be used
f.to_csv("../cleaned_data/by_county_all_merged.csv") 
```
