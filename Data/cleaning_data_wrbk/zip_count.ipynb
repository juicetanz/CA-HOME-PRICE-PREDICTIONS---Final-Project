```python
# Dependencies and Setup
import pandas as pd
import pathlib
import csv
import scipy.stats as st
import numpy as np
import calendar
from time import strptime
```

```python
# Create paths for csv
zip_county_path = "../raw_data/zip_county.csv"
```

```python
# Read csv
zip_county_df = pd.read_csv(zip_county_path)
pd.DataFrame(zip_county_df)
```

```python
# remove all states that are not CA (California)
only_CA_df = zip_county_df.loc[(zip_county_df['STATE'] == 'CA' )]
only_CA_df
```

```python
# Delete extraneous columns
delete_extraneous_columns = only_CA_df.drop(columns= [
    "STATE"
    , "STCOUNTYFP"
    , "CLASSFP"
])    
delete_extraneous_columns.head()
```

```python
# Rename columns
new_zip_county_df = delete_extraneous_columns.rename(columns={"ZIP":"zip", "COUNTYNAME": "countyName"})
new_zip_county_df
```

```python
# Remove the county from each name
new_zip_county_df["countyName"] = new_zip_county_df["countyName"].str.replace(" County", "")
new_zip_county_df[['countyName', "zip"]]
```

```python
# Finding the number of counties in this dataset
new_zip_county_df["countyName"].nunique()
```

```python
# Save as csv
new_zip_county_df.to_csv("../cleaned_data/zip_county.csv", index=False)
```
