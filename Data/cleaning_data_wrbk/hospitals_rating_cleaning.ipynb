```python
# Dependencies and Setup
import pandas as pd
import pathlib
import csv
import scipy.stats as st
import numpy as np
import calendar
from time import strptime
```

```python
# Create paths for csv
hospital_rating = "../raw_data/hospital_dataset/hospital_rating.csv"
```

```python
# Read csv
hosptial_rating_df = pd.read_csv(hospital_rating)
pd.DataFrame(hosptial_rating_df)
```

```python
# remove all states that are not CA (California)
only_CA_df = hosptial_rating_df.loc[(hosptial_rating_df['State'] == 'CA' )]
only_CA_df
```

```python
# List out all the columns names to determine what compenents are necessary 
only_CA_df.columns.tolist()
```

```python
# Delete extraneous columns
delete_extraneous_columns = only_CA_df.drop(columns= [
    "Facility ID"
    , "Facility Name"
    , "Address"
    , "City"
    , "State"
    , "Phone Number"
    , "Hospital Type"
    , "Emergency Services"
    , "Meets criteria for promoting interoperability of EHRs"
    , "Hospital overall rating footnote"
    , "Hospital Ownership"
    , "READM Group Footnote"
    , "Pt Exp Group Footnote"
    , "TE Group Footnote"
    , "Safety Group Measure Count" 
    , "Meets criteria for promoting interoperability of EHRs"
    , "Hospital overall rating footnote"
    , "MORT Group Footnote"
    , "Safety Group Footnote"
    , "Count of Facility Safety Measures"
    , "Count of Safety Measures Better"
    , "Count of Safety Measures No Different"
    , "Count of Safety Measures Worse"
    , "Pt Exp Group Measure Count"
    , "Count of Facility Pt Exp Measures"
    , "TE Group Measure Count"
    , "Count of Facility TE Measures"
    , "Count of Facility MORT Measures"
    , "Count of MORT Measures Better"
    , "Count of MORT Measures No Different"
    , "Count of MORT Measures Worse"
    , "Count of Facility READM Measures"
    , "Count of READM Measures Better"
    , "Count of READM Measures No Different"
    , "Count of READM Measures Worse"
    , "MORT Group Measure Count"
    , "READM Group Measure Count"
    , "ZIP Code"
])    
delete_extraneous_columns.head()
```

```python
# Rename columns
new_hosptial_rating_df = delete_extraneous_columns.rename(columns={"County Name":"countyName", "Hospital overall rating": "hospitalOverallRating"})
new_hosptial_rating_df
```

```python
# Finding the number of counties in this dataset
new_hosptial_rating_df["countyName"].nunique()
```

```python
# Version 1 - Dataframes excluding "Not Available" Hospital overall rating
vr1_without_na_df = new_hosptial_rating_df.loc[(new_hosptial_rating_df['hospitalOverallRating'] != 'Not Available' )]
vr1_without_na_df
```

```python
# Finding the number of counties without the "Not Available" status
vr1_without_na_df.groupby("countyName").sum().shape
```

```python
# Save as csv
vr1_without_na_df.to_csv("../cleaned_data/hospitalrating_vr1_df.csv", index=False)
```

```python
# Version 2 - substituting zeros
vr2_sub_0_df = new_hosptial_rating_df.replace("Not Available", 0)
pd.DataFrame(vr2_sub_0_df)
```

```python
# Save Version 2 Hospital rating to csv
vr2_sub_0_df.to_csv("../cleaned_data/hospitalrating_vr2_df.csv", index=False)
```

```python
# Version 3 - Replacing video adding median/average
vr3_med = vr1_without_na_df["hospitalOverallRating"].astype(int).median()
vr3_med
```

```python
# Substite the "Not Available" with the median/average 
ver3_df = new_hosptial_rating_df.replace("Not Available", 3)
pd.DataFrame(ver3_df)
```

```python
# Final cleaned csv to be used
ver3_df.to_csv("../cleaned_data/hospitalrating_vr3_df.csv", index=False)
```

```python
# Create a new variable to remember the call
ver3_df.dtypes
```

```python
# Group by County
by_county_hospitalrating_df = ver3_df.astype({"countyName":str,"hospitalOverallRating":int}).groupby("countyName").mean() 
pd.DataFrame(by_county_hospitalrating_df)
```

```python
# Final cleaned csv to be used
by_county_hospitalrating_df.to_csv("../cleaned_data/by_county_hospitalrating_ver3_df.csv")
```
